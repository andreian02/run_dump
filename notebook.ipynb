```python
import glob
import json
import datetime
import pandas as pd
```

```python
alldata = glob.glob("*.json")
len(alldata)
```

```python
# loop.
allframes = []

list_1 = []
list_2 = []

for item in alldata:
    with open(item) as f1:
        data1 = json.load(f1)
        var_id = data1['id'] 
        epochtime= (data1['start_epoch_ms']/1000)
        date = datetime.datetime.fromtimestamp(epochtime).date()
        date_run = datetime.datetime.fromtimestamp(epochtime)
        
        #print(date_run)
        #process data['tags']
        
        payload1 = data1['tags']
        payload1['date'] = date
        payload1['start_date'] = date_run
        payload1['v_id'] = var_id
        print(payload1)
        

       
       
        allframes.append(payload1)

```

```python
frames = []


for item in alldata:
    with open(item) as f1:
        data1 = json.load(f1)
        dict_1 = {}
        var_id = data1['id'] 
        start = data1['start_epoch_ms'] / 1000.0
        end = data1['end_epoch_ms'] / 1000.0
        s_time = (end - start) / 60
        
        dt_s = dt = datetime.datetime.fromtimestamp(start).strftime('%Y-%m-%d %H:%M:%S.%f')
        dt_e = dt = datetime.datetime.fromtimestamp(end).strftime('%Y-%m-%d %H:%M:%S.%f')
    
        for x in data1['summaries']:
            print(x['metric'], ":", x['value'])
            
            dict_1[x['metric']] = x['value']
            dict_1['v_id'] = var_id
            dict_1['start'] = dt_s
            dict_1['end'] = dt_e
            dict_1['time_mins'] = s_time
            
        #frames.append(dict_1)
        print("-"*10)
        print(dict_1)
        print("-"*10)
        frames.append(dict_1)
        
        
allrun = pd.DataFrame(frames)     
```
